```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('cleaned_data.csv')
```

```python
#POSTED_BY - Pie chart to represent percentage
#UNDER_CONSTRUCTION -percentage
#RERA - percentage
#BHK_NO - Bar chart
#BHK_OR_RK - Percentage
#READY_TO_MOVE _Percentage

#RESALE -bar chart
#CITY - bar chart
```

```python
posted_by = ds.groupby("POSTED_BY")["TARGET(PRICE_IN_LACS)"].count()
posted_by.plot(kind= 'pie',autopct= '%1.1f%%', startangle=90, figsize=(6,6))
```

```python
under_construction = ds.groupby("UNDER_CONSTRUCTION")["UNDER_CONSTRUCTION"].count()
under_construction
```

```python
under_construction.plot(kind= 'pie',autopct= '%1.1f%%', startangle=90, figsize=(6,6))
```

```python
rera = ds.groupby("RERA")["RERA"].count()
rera
rera.plot(kind='pie', autopct= '%1.1f%%')
```

```python
bhk = ds.groupby("BHK_NO.")["BHK_NO."].count()
ax = bhk.plot(kind='bar', color= 'pink', figsize=(8,6))
ax.legend(["Count of Entries"], loc="upper right")
# Set Lables and titile 
plt.xlabel("Number of BHK")
plt.ylabel("Count of Entries")
bhk = ds.groupby("BHK_NO.")["BHK_NO."].count()
ax = bhk.plot(kind='bar', color= 'pink', figsize=(8,6))
ax.legend(["Count of Entries"], loc="upper right")
for i, v in enumerate (bhk):
    ax.text(i, v+0.1, str(v), ha='center', va= 'bottom', fontsize= 10)
#Display the plot 
plt.show()
```

```python
citywisecount = ds.groupby("CITY")["CITY"].count()
citywisecount = citywisecount.sort_values(ascending=False)
counts = citywisecount.head(10)
```

```python
# Group by "City" and count the entries 
citywisecount = ds.groupby("CITY")["TARGET(PRICE_IN_LACS)"].count()

# Oreder tjhe result in descending order
citywisecount = citywisecount.sort_values(ascending=False)

#Get top 10 cities
counts = citywisecount.head(10)
counts_df = pd.DataFrame(counts).reset_index()
counts_df

#plot using Seaborn (Horizantal bar Chart
plt.figure(figsize=(10,6))
sns.barplot(data=counts_df, x="CITY",y="TARGET(PRICE_IN_LACS)",palette="Blues_d")

#set labels and title
plt.xlabel("City")
plt.ylabel("Count of Entries")
plt.title("Top 10 Cities by Count")  

#Display the plot
plt.xticks(rotation=45) #Rotate X-axis Lables for better visiblity
plt.show()

```
